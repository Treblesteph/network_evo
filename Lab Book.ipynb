## Learning how to draw with compose

```julia
@manipulate for bend1 = -1:0.01:1, bend2 = -1:0.01:1,
                bend3 = -1:0.01:1, bend4 = -1:0.01:1,
                bend5 = -1:0.01:1, bend6 = -1:0.01:1,
                bend7 = -1:0.01:1, bend8 = -1:0.01:1
    compose(context(),
    (context(), polygon([(0.39, 0.39), (0.4, 0.45), (0.45, 0.4)]),
                fill(LCHab(8, 258, 290))),
    (context(), curve([(0.19, 0.3), (0.15, 0.15)],
                [(bend1, bend2), (bend3, bend4)],
                [(bend5, bend6), (bend7, bend8)],
                [(0.15, 0.15), (0.3, 0.19)]),
                fill(nothing), linewidth(1mm),
                Compose.stroke(LCHab(8, 258, 290))),
    (context(), circle(0.3, 0.3, 0.1), fill(LCHab(88, 84, 219))),
    (context(), circle(0.7, 0.7, 0.1), fill(LCHab(88, 84, 219))),
    (context(), circle(0.7, 0.3, 0.1), fill(LCHab(88, 84, 219))),
    (context(), circle(0.3, 0.7, 0.1), fill(LCHab(88, 84, 219))),
    (context(), line([(0.4, 0.4), (0.6, 0.6)]), linewidth(1mm),
                Compose.stroke(LCHab(8, 258, 290))),
    (context(), Compose.rectangle(0, 0, 1, 1),
                fill(LCHab(68, 144, 182))))
end
```

# 22/09/2014: Plotting concentration timeseries with day shading.

```julia
using HDF5, JLD
data = load("out_2014-09-19_11_05.jld")
```

```julia
include("evolveclock.jl")
import(clockga)

concsFrame1 = DataFrame()
concsFrame1[:time] = (1:4*24*60) / 60
concsFrame1[:gene1] = data["concs1"][:, 1]
concsFrame1[:gene2] = data["concs1"][:, 2]
concsFrame1[:gene3] = data["concs1"][:, 3]
concsFrame1[:gene4] = data["concs1"][:, 4]
concCombinations = (concsFrame1[:, [1:2]],
                    concsFrame1[:, [1, 3]],
                    concsFrame1[:, [1, 4]],
                    concsFrame1[:, [1, 5]],
                    concsFrame1[:, [1:3]],
                    concsFrame1[:, [1:2, 4]],
                    concsFrame1[:, [1:2, 5]],
                    concsFrame1[:, [1, 3:4]],
                    concsFrame1[:, [1, 3, 5]],
                    concsFrame1[:, [1, 4:5]],
                    concsFrame1[:, [1:4]],
                    concsFrame1[:, [1:3, 5]],
                    concsFrame1[:, [1:2, 4:5]],
                    concsFrame1[:, [1, 3:5]],
                    concsFrame1[:, [1:5]])
colour1 = "mediumturquoise"
colour2 = "orchid"
colour3 = "dodgerblue"
colour4 = "coral"
colours = [colour1, colour2, colour3, colour4]
genes = [:gene1, :gene2, :gene3, :gene4]
@manipulate for i in 1:length(concCombinations)
    concsSubFrame1 = stack(concCombinations[i], 2:ncol(concCombinations[i]))
    plot(layer(concsSubFrame1, x = "time", y = "value",
               color = "variable",
               Scale.discrete_color_manual(colours..., levels = genes),
               Geom.line,
               Scale.x_continuous(minvalue=0, maxvalue=96),
               Guide.xticks(ticks=[0:12:96]), Guide.yticks(ticks=[0,1])),
         layer(xmin = clockga.DAWNS[:, 1],
               xmax = clockga.DAWNS[:, size(clockga.DAWNS,2)],
               y = ones(size(clockga.DAWNS,1)),
               color = repeat(["Dawns"], outer=[4]),
               Scale.discrete_color_manual("cornsilk"),
               Geom.bar))
end
```
