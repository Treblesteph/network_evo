# 22/09/2014: Plotting concentration timeseries with day shading.

### First generating timeseries to look at different gene combinations separately.

```julia
using HDF5, JLD
data = load("out_2014-09-19_11_05.jld")
```

```julia
include("evolveclock.jl")
import(clockga)

function plotconcs(concsdata)
    concsFrame1 = DataFrame()
    concsFrame1[:time] = (1:4*24*60) / 60
    concsFrame1[:gene1] = concsdata["concs1"][:, 1]
    concsFrame1[:gene2] = concsdata["concs1"][:, 2]
    concsFrame1[:gene3] = concsdata["concs1"][:, 3]
    concsFrame1[:gene4] = concsdata["concs1"][:, 4]
    concCombinations = (concsFrame1[:, [1:2]],
                        concsFrame1[:, [1, 3]],
                        concsFrame1[:, [1, 4]],
                        concsFrame1[:, [1, 5]],
                        concsFrame1[:, [1:3]],
                        concsFrame1[:, [1:2, 4]],
                        concsFrame1[:, [1:2, 5]],
                        concsFrame1[:, [1, 3:4]],
                        concsFrame1[:, [1, 3, 5]],
                        concsFrame1[:, [1, 4:5]],
                        concsFrame1[:, [1:4]],
                        concsFrame1[:, [1:3, 5]],
                        concsFrame1[:, [1:2, 4:5]],
                        concsFrame1[:, [1, 3:5]],
                        concsFrame1[:, [1:5]])
    colour1 = "mediumturquoise"
    colour2 = "orchid"
    colour3 = "dodgerblue"
    colour4 = "coral"
    colours = [colour1, colour2, colour3, colour4]
    genes = [:gene1, :gene2, :gene3, :gene4]
    @manipulate for i in 1:length(concCombinations)
        concsSubFrame1 = stack(concCombinations[i], 2:ncol(concCombinations[i]))
        plot(concsSubFrame1, x = "time", y = "value",
             color = "variable",
             Scale.discrete_color_manual(colours..., levels = genes),
             Geom.line,
             Scale.x_continuous(minvalue=0, maxvalue=96),
             Guide.xticks(ticks=[0:12:96]), Guide.yticks(ticks=[0,1]))
    end
end

plotconcs(data)
```

### Making dawn and dusk shading.

```julia
include("evolveclock.jl")
import(clockga)
# Dawns
DAYS = clockga.DAYS
DAWNS = clockga.DAWNS
DUSKS = clockga.DUSKS
DAYS_START = (DAWNS[:, size(DAWNS,2)] + 1) / 60
DAYS_END = DAYS[:, size(DAYS,2)] / 60
NIGHTS_START = DAYS_END + 1/60
NIGHTS_END = 24 + DAYS_START - 1/60
DAWNS_START = DAWNS[:, 1] / 60
DAWNS_END = DAWNS[:, size(DAWNS,2)] / 60
DUSKS_START = DUSKS[:, 1] / 60
DUSKS_END = DUSKS[:, size(DUSKS,2)] / 60
plot(xmin = [DAWNS_START, DUSKS_START, DAYS_START, NIGHTS_START],
     xmax = [DAWNS_END, DUSKS_END, DAYS_END, NIGHTS_END],
     Scale.x_continuous(minvalue=0, maxvalue=96),
     y = ones(size(DAWNS,1) + 
              size(DUSKS,1) +
              size(DAYS,1) +
              size(DAYS,1)),
     color = repeat(["Dawns", "Dusks", "Days", "Nights"], inner=[4]),
Scale.discrete_color_manual("palegoldenrod", "lightgrey", "cornsilk", "azure2"),
     Guide.xticks(ticks=[0:12:96]), Guide.yticks(ticks=[0,1]),
     Geom.bar)
```
